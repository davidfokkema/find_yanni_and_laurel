```python
from IPython.display import Audio
```

```python
import librosa
```

```python
low, high = -10, 10
```

```python
def get_single_word_from_audio(audio):
    intervals = librosa.effects.split(audio, top_db=20)
    start, end = intervals[0]
    return audio[start:end]
```

```python
def get_audio_shift_function(filename):
    full_audio, rate = librosa.load(filename)
    audio = get_single_word_from_audio(full_audio)

    def play_pitch_shifted(shift):
        shifted = librosa.effects.pitch_shift(audio, rate, shift)
        display(Audio(shifted, rate=rate, autoplay=True))

    return play_pitch_shifted
```

```python
play_pitch_shifted = get_audio_shift_function('Yanni_or_Laurel.m4a')
```

```python
def verify_name_for_shift(shift, names):
    play_pitch_shifted(shift)
    name_heard = input("Did you hear Yanni or Laurel?")
    if name_heard.lower() not in names:
        raise RuntimeError("You heard %s, but it should be one of %s" % (name_heard, names))
```

```python
yanni, laurel = ['y', 'yanni'], ['l', 'laurel']
```

```python
verify_name_for_shift(low, yanni)
```

```python
verify_name_for_shift(high, laurel)
```

```python
def find_perfect_pitch(low, high):
    while True:
        mid = (low + high) / 2
        print("Playing with pitch shifted %.1f halftones" % mid)
        play_pitch_shifted(mid)
        name_heard = input("Did you hear Yanni or Laurel?").lower()
        if name_heard in yanni:
            low = mid
        elif name_heard in laurel:
            high = mid
        else:
            return
```

```python
find_perfect_pitch(low, high)
```
